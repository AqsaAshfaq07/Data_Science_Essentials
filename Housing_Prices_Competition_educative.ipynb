# DATA PREPROCESSING

```python
housing.isnull().sum().sort_values(ascending = False).head(15)
```

```python
#DEALING WITH MISSING DATA
```

```python
housing_processed = housing.copy()

cate_cols = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu',
                     'GarageCond', 'GarageQual', 'GarageFinish', 'GarageType',
                     'BsmtFinType2', 'BsmtExposure', 'BsmtFinType1', 'BsmtQual', 'BsmtCond',
                     'MasVnrType']
cate_cols

# Replace missing values in categorical columns with None
for cat in cate_cols :
    housing_processed[cat] = housing_processed[cat].fillna('None')

#Replace lotfrantage missing values with median of the neighborhood column values
housing_processed['LotFrontage'] = housing_processed.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x : x.fillna(x.median()) )

#Replacement in Garage....
for col in ['GarageYrBlt', 'GarageArea', 'GarageCars', 'MasVnrArea'] :
    housing_processed[col] = housing_processed[col].fillna(int(0))
    
#Replacement in Electrical column
housing_processed['Electrical'] = housing_processed['Electrical'].fillna(housing_processed['Electrical']).mode()[0]
    
#There is no need of Utilities column -- so let's just drop it out 
housing_processed = housing_processed.drop(['Utilities'], axis = 1)

#NOW GET THE COUNT TO CHECK WE ARE LEFT WITH NO MISSING VALUES

housing_processed.isnull().sum()
#OR use this method 
housing_processed.isnull().apply(sum).max()
```

```python
#DEALING WITH OUTLIERS
```

```python
num_data = housing_processed.select_dtypes(exclude = 'object')  # housing_processed --> data with no missing values
high_quant = housing_processed.quantile(.999)  
```

```python
for i in num_data.columns:   
    housing_processed  = housing_processed.drop(housing_processed[i][housing_processed[i]>high_quant[i]].index)  # .index --> specific row labels
    
housing_processed.info()
```

```python
#Remove highly correlated features
drop_attributes = ['MiscVal', 'MoSold', 'YrSold', 'BsmtFinSF2','BsmtHalfBath','MSSubClass',
                   'GarageArea', 'GarageYrBlt', '3SsnPorch']
housing_processed.drop(drop_attributes, axis = 1).shape
```

DATA TRANSFORMATION

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer  #handles both numerical and categorical data values
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import StandardScaler

#separate features and target variable 
housing_X = housing_processed.drop('SalePrice', axis=1)  # input  -- remove the target feature from trainig dataset
housing_y = housing_processed['SalePrice'].copy()  # output

#separate out numerical and categorical variables
num_vars = housing_X.select_dtypes(exclude = 'object')
cate_vars = housing_X.select_dtypes(include = 'object')

numerical = list(num_vars)
categorical = list(cate_vars)

#Applying pipeline to impute any missing values inside the training dataset
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = "median")),
    ('std_scaler', StandardScaler())
])
num_pipeline
```

```python
#CREATING THE PIPELINE FOR ALL DATA GENRES  -- using columntransformer
full_pipeline = ColumnTransformer([
    ('num', num_pipeline, numerical),
    ('cat', OneHotEncoder(), categorical)
])

#print and describe it before transformation 
print(housing_y.describe())

#apply log transformation to make data more normally distributed
housing_y_prepared = np.log(housing_y)

#Run the transformation on input data set
housing_X_prepared = full_pipeline.fit_transform(housing_X)

# description before applying transform
print(housing_y_prepared.describe())

housing_X_prepared
```
