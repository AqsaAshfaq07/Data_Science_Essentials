```python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)
```

```python
#SCATTER PLOT
```

```python
plt.scatter(x, y, c= colors, s= sizes, alpha = 0.9, cmap= plt.cm.Purples) #alpha --> thickness of the dots...// transparency level
#cmap --> Color map ... specifies the color of dots 
#size --> this argument is in pixels
plt.colorbar() # to also add the color scale
```

```python
num1 = data1['Country/Territory']
num2 =  data1['2022 Population']
plt.bar(num1, num2, align='center', alpha= 0.7)
plt.xlabel('Country')
plt.ylabel('2022')
plt.title('Population Data for the year 2022')
```
